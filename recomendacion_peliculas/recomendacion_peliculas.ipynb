```python
import pandas as pd
```

```python
df = pd.read_csv("movie_metadata.csv")

df
```

```python
df["genero"] = df["genero"].str.replace("|", " ")
df
```

```python
df["plot_keywords"] = df["plot_keywords"].str.replace("|", " ")
df
```

```python
df["texto"] = df[["genero", "plot_keywords"]].apply(lambda row: " ".join(row.values.astype(str)), axis=1)
```

```python
df
```

En Pandas, la función iloc se utiliza para seleccionar filas y columnas de un DataFrame basándose en su posición numérica (índice entero).

```python
row = df[["genero", "plot_keywords", "texto"]].iloc[0]
print(row)
```

```python
print(df["texto"].iloc[0])
```

TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
```

```python
tfidf = TfidfVectorizer(max_features=2000)
```

```python
#matriz de vectores
X = tfidf.fit_transform(df["texto"])
```

```python
X
```

```python
peliculas = pd.Series(df.index, index=df["movie_title"])
```

```python
peliculas
```

La función strip() en Python sirve para eliminar espacios en blanco o caracteres específicos al inicio y al final de una cadena de texto.

```python
#para eliminar los espacios del principio y del final
peliculas.index = peliculas.index.str.strip()
peliculas
```

```python
indice = peliculas["The Dark Knight Rises"]
indice
```

CONSULTA

```python
consulta = X[indice]
consulta
```

```python
consulta.toarray()
```

Calcular las similitudes de los vectores por coseno

```python
similitud = cosine_similarity(consulta, X)
similitud
```

```python
similitud[0][3]
```

```python
similitud = similitud.flatten()
similitud[3]
```

Ordenando las similitudes de la pelicula que seleccione de mayor a menor

```python
(-similitud).argsort()
```

```python
plt.plot(similitud[(-similitud).argsort()])
```

Ahora una recomendacion de las 10 peliculas mas parecidas a la que selecione

```python
recomendacion = (-similitud).argsort()[1:11]
recomendacion #guarda los indices no mas
```
